Heart Disease prediction with Logistic Regression


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

```python
data=pd.read_csv("/content/heart_disease_data.csv")
```

```python
x=data.drop(columns='target',axis=1)
y=data['target']
```

```python
x
```

```python
y
```

DATA SPLITING


```python
x_train , x_test ,y_train , y_test=train_test_split(x,y,test_size=0.25,stratify=y,random_state=2)
```

```python
x.shape ,x_train.shape, x_test.shape
```

MODEL TRAINING

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
training_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
"Accuracy on training data : ", training_accuracy
```

```python
input_data = (62,0,0,140,268,0,0,160,0,3.6,0,2,2)
input_numpy_array= np.asarray(input_data)
input_reshaped = input_numpy_array.reshape(1,-1)
```

```python
prediction=model.predict(input_reshaped)
print(prediction)

if (prediction[0]==0):
  print('Person does not have heat disease. ')
else:
  print('Person has heart disease.')
```
